```python
import mido as m
from mido import MidiFile

mid = MidiFile("samples\\my_heart_will_go_on1.mid")
```

```python
for i, track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i, track.name))
    for msg in track:
        print(msg)
```

```python
for i, track in enumerate(mid.tracks):
    if track.name == "Synth Voice":
        i = 0
        for msg in track:
            if(i < 10):
                print(msg)
                print(msg.hex())
                i = i + 1
```

```python
pan_flute = MidiFile("samples\\Pan Flute 3.mid")
for i, track in enumerate(pan_flute.tracks):
    i = 0
    for msg in track:
        i = i + 1
        if(i < 500):
            if(msg.type == 'note_on' or msg.type == 'note_off'):
                print(msg)
                print(msg.hex())
        
```
